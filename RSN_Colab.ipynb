```python
# Download images from GitLab
# Copy pics link from GitLab
!wget https://gitlab.umiacs.umd.edu/vishnuds/cmsc421/-/raw/main/pics.zip

# Unzip images
# Images are in Files/content/pics
!unzip pics.zip

# Run yoloV5 from pytorch for the images

# Install yolov5
!pip install -qr https://raw.githubusercontent.com/ultralytics/yolov5/master/requirements.txt  # install dependencies

# Import pyTorch
import torch

# Import glob package
import glob

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Images
imgs = ['https://ultralytics.com/images/zidane.jpg']  # batch of images
imgs = glob.glob('pics/*.jpeg')  # batch of images
# Inference
results = model(imgs)
print(dir(results))

# Results
results.print()
results.save()  # or .show()


results.xyxy[0]  # img1 predictions (tensor)
results.pandas().xyxy[0]  # img1 predictions (pandas)
#      xmin    ymin    xmax   ymax  confidence  class    name
# 0  749.50   43.50  1148.0  704.5    0.874023      0  person
# 1  433.50  433.50   517.5  714.5    0.687988     27     tiey
# 2  114.75  195.75  1095.0  708.0    0.624512      0  person
# 3  986.00  304.00  1028.0  420.0    0.286865     27     tie
```
